```python
##Import Librarier
```

```python
import pandas as pd
import os
```

```python


files = [f for f in os.listdir('./Sales_Data')]
all_data = pd.DataFrame()

for f in files:
    df = pd.read_csv(f"./Sales_Data/{f}")
    all_data = pd.concat([all_data,df])
all_data.to_csv("all_data.csv", index=False)

all_data.head()
```

```python
###Cleaning Data
```

```python
all_data = all_data.dropna()
all_data.head()
```

```python
###Finding Sales Per Month and which month had the most Sales
```

```python
###Month Column
```

```python
all_data['Month'] = all_data['Order Date'].str[0:2]
#filter or"
all_data = all_data[all_data['Order Date'].str[0:2] !='Or']

all_data['Month'] = all_data['Month'].astype('int32')
all_data.head()
```

```python
#Convert to int and float
all_data['Quantity Ordered']= pd.to_numeric(all_data['Quantity Ordered'])
all_data['Price Each'] =pd.to_numeric(all_data['Price Each'])
all_data.head()
```

```python
###Sales column
```

```python
all_data['Sales'] = all_data['Quantity Ordered']*all_data['Price Each']
all_data.head()
```

```python
###sales in a month
```

```python
results = all_data.groupby('Month').sum()
#all_data.groupby('Month').sum()['Sales'].sort_values(ascending=0)
```

```python
import matplotlib.pyplot as plt
```

```python
months = range(1,13)
plt.bar(months, results['Sales'])
plt.xticks(months)
plt.ylabel('Sales in USD ($)')
plt.xlabel('Month Number')
plt.show()
```

```python
#finding city with highest number of sales
```

```python
all_data.head()
```

```python
###Add City column

all_data['City']= all_data['Purchase Address'].apply(lambda x : x.split(',')[1].strip())
all_data.head()
```

```python
results =all_data.groupby('City').sum()
#all_data.groupby('City').sum()['Sales'].sort_values(ascending=0)
```

```python
city = [city for city, df in all_data.groupby('City')]
plt.bar(city, results['Sales'])
plt.xticks(city, rotation ='vertical', size=8)
plt.ylabel('Sales in USD ($)')
plt.xlabel('City')
plt.show()
```

```python
### what time should we display advertisements to maximize sales
```

```python
all_data['Order Date'] = pd.to_datetime(all_data['Order Date'])
```

```python
all_data.head()
```

```python
all_data['Hour'] = all_data['Order Date'].dt.hour
all_data['Minute'] = all_data['Order Date'].dt.minute

all_data.head()
```

```python
all_data.groupby('Hour').count()
```

```python
hours = [hour for hour, df in all_data.groupby('Hour')]
plt.plot(hours, all_data.groupby('Hour').count())
plt.xticks(hours)
plt.xlabel('Hour')
plt.ylabel('Number of Orders')
plt.grid()
plt.show()
```

```python
#just before noon(around lunch time) and around 7 pm
```

```python
minutes = [minute for minute, df in all_data.groupby('Minute')]
plt.plot(minutes, all_data.groupby('Minute').count())
```

```python
###What Products are most often sold together
```

```python
all_data.head()
```

```python
Products = [product for product, df in all_data.groupby('Product')]
print(Products)
```

```python

df = all_data[all_data['Order ID'].duplicated(keep=False)]
df.head()
```

```python
df['All Products'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
df.head()
#df.groupby(['Order ID','Product']).count()
```

```python
df = df[['Order ID', 'All Products']].drop_duplicates()
df.head()
```

```python
from itertools import combinations
from collections import Counter
```

```python
count = Counter()

for row in df['All Products']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list,2)))

for key,value in count.most_common(10):
    print(key,value)
```

```python
product_grouped = all_data.groupby('Product')['Quantity Ordered']
quantity_ordered=product_grouped.sum()
```

```python
products = [product for product, df in all_data.groupby('Product')]
plt.bar(products, quantity_ordered)
plt.xticks(products, rotation ='vertical', size=8)
plt.xlabel('Products')
plt.ylabel('Quantity Ordered')
plt.show()
```

```python
prices = all_data.groupby('Product')['Price Each']
prices_sumed = prices.mean()
print(prices_sumed)
```

```python
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.bar(products,quantity_ordered, color = 'r')
ax2.plot(products,prices_sumed,'b-')

ax1.set_xlabel('Product Name')
ax1.set_ylabel('Quantity Ordered', color = 'r')
ax2.set_ylabel('Price', color = 'b')
ax1.set_xticklabels(products, rotation ='vertical', size =8)

plt.show()
```
