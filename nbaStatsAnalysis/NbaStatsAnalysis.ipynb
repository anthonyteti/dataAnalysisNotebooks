```python
import pandas as pd
```

```python
import chardet


df = pd.read_csv('nba_stats_database.csv')
df.head()
```

```python
#Calculate per game columns

df['PPG'] = (pd.to_numeric(df['PTS'], errors='coerce'))/(pd.to_numeric(df['G'], errors='coerce'))
df['APG'] = (pd.to_numeric(df['AST'], errors='coerce'))/(pd.to_numeric(df['G'], errors='coerce'))
df['RPG'] = (pd.to_numeric(df['TRB'], errors='coerce'))/(pd.to_numeric(df['G'], errors='coerce'))
df['3PG'] = (pd.to_numeric(df['3PA'], errors='coerce'))/(pd.to_numeric(df['G'], errors='coerce'))
df['BPG'] = (pd.to_numeric(df['BLK'], errors='coerce'))/(pd.to_numeric(df['G'], errors='coerce'))
df['SPG'] = (pd.to_numeric(df['STL'], errors='coerce'))/(pd.to_numeric(df['G'], errors='coerce'))
df.round(3)
```

```python
pd.set_option('display.max_columns', None)
print(df)
```
